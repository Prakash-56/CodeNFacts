```python
import json
```

```python
#Load the data
def load_data(filename):
    with open (filename, "r") as f:
        data=json.load(f)
        
    return data
```

```python
data=load_data("store-data.json")
print(data)
print(type(data))
```

```python
#clean & structure the data
def clean_data(data):
    text_to_num = {"one":1, "two":2, "three":3, "four":4, "five":5}
    
    for user in data:
        #clean ratings
        raw_rating = user["rating"].strip().lower()
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
    
        print(raw_rating)
    
```
